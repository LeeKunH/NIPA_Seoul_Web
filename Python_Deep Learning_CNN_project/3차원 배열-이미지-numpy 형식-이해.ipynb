##### Q. 0번 차원 인덱스는 빨간색 층을 의미한다고 했음. 근데 출력을 해보면 왜 초록색이 뜨는거지?
내가 예상한 결과는 R 0 이면 해당 픽셀 빨간색 성분이 없다는 의미 > 검정색 출력, R이 255면 가장 진한 빨간색 출력됨.
따라서 전체적으로 빨간색이 출력되야 함

ㅡ 이유설명 =>

빨간색 채널값을 시각화하면 출력 이미지가 초록색 계열로 보이는 이유는 이미지 시각화 라이브러리 (예: Matplotlib)에서 이미지를 표시할 때 RGB 채널의 순서를 고려하지 않았기 때문일 수 있음.

즉, 현재 채널은 빨간색 채널인데 라이브러리에서 이 순서를 고려하지 않아서 발생한 것

따라서 " 현재 출력하고자 하는 층은 빨간색 채널 층이니 빨간색으로 강도 또는 밝기 포현해줘 " 를 알려줘야함

Matplotlib에서는 `cmap` 매개변수를 사용하여 채널 순서를 지정할 수 있음

```python
plt.imshow(img[:, :, 0])
```

```python
plt.imshow(img[:, :, 0],cmap='Reds')
```

```python
plt.imshow(img[:, :, 1])
```

```python
plt.imshow(img[:, :, 1], cmap='Greens')
```

```python
plt.imshow(img[:, :, 2])
```

```python
plt.imshow(img[:, :, 2], cmap='Blues')
```

ㅡ
인공지능 모델
학습 성능을 올리기 위해
0~255값 범위를 가지고 있던 배열값들을 / 255를 진행 = 노멀라이즈 진행.


ㅡ
노멀라이즈"란 데이터의 범위를 조정하여 모든 값들을 일정한 범위 내에 맞추는 작업을 의미합니다. 이것을 비유로 설명하면, 예를 들어 과일 바구니 안에 사과와 바나나가 있을 때, 사과와 바나나의 크기를 비슷하게 맞추는 것과 비슷합니다. 그래서 노멀라이즈된 데이터는 모든 값이 일정한 범위 내에 있어서 모델이 더 쉽게 학습하고 판단할 수 있습니다.

```python
img_array = img / 255
img_array
```

ㅡ
img/255를 했음에도
원본 이미지가 그대로 출력됨을 확인 가능. = 원본 정보를 잃어버리지 않는다.

```python
plt.imshow(img_array)
```
